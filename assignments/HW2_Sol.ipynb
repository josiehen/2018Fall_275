In the $H_2^+$ molecule, there are two nuclei and one electron. The Hamiltonian is

$$H =-\frac1 2 \Delta + V_{ext}(\bvec r, \{\bvec R_1\}) + V_{ext}(\bvec r, \{\bvec R_2\}) + \frac{1}{\vert \bvec R_1-\bvec R_2\vert }$$

The spin degree of freedom can be factored out.  The Kohn-Sham energy functional is


$$\mathcal E ^{KS} (\psi) = \frac{1}{2} \int \vert  \nabla \psi(\bvec r) \vert^{2} \ud \bvec r + \frac 1 2 \int \frac{\rho(\bvec r) \rho(\bvec r')}{\vert \bvec r-\bvec r' \vert}\ud \bvec r \ud \bvec r' + E_{xc}[\rho] + \int \rho(r) (V_{ext} (\bvec r, \bvec R_{1}) + V_{ext} (\bvec r, \bvec R_{2}) ) \ud \bvec r +  \frac{1}{\vert \bvec R_1-\bvec R_2\vert }$$


The Euler-Lagrange equation as 
$$\left(-\frac1 2 \Delta + V_{ext}(\bvec r, \{\bvec R_1\}) + V_{ext}(\bvec r, \{\bvec R_2\})  + \int \frac{\rho(\bvec r')}{\vert  \bvec r-\bvec r' \vert}\ud \bvec r' + V_{xc}[\rho]\right) \psi(\bvec r) = \epsilon \psi(\bvec r).$$




2)


According to the definition of the Wannier function $w_{n, \bvec R} (\bvec r)$

$$w_{n, \bvec R} (\bvec r) = \frac{1}{\vert \Omega^*\vert}\int_{\Omega^\ast} \tilde \psi_{n, \bvec k} (\bvec r) e^{-i\bvec k \cdot \bvec R}\ud \bvec k,
$$

we have
$$\int_{\RR^3}w_{n, \bvec R}^*(\bvec r)w_{n', \bvec R'}(\bvec r) \ud \bvec r = \frac{1}{\vert \Omega^* \vert ^2} \int _{\RR^3} \int _{\Omega^*}\int _{\Omega^*} \tilde \psi_{n, \bvec k}^* (\bvec r) \tilde \psi_{n', \bvec k'} (\bvec r) e^{i\bvec k \cdot \bvec R -i\bvec k' \cdot \bvec R'}\ud \bvec k \ud \bvec k'\ud \bvec r. $$

Given the definition
$$\tilde \psi_{n', \bvec k'} (\bvec r) = \sum_{p}\psi_{p, \bvec k} (\bvec r) U_{p,n}(\bvec k),$$

the left hand side can be rewritten as
\begin{eqnarray}
\int_{\RR^3}w_{n, \bvec R}^*(\bvec r)w_{n', \bvec R'}(\bvec r) \ud \bvec r &=& \frac{1}{\vert \Omega^* \vert ^2} \int _{\RR^3} \int _{\Omega^*}\int _{\Omega^*}  \sum_{p,q=1}^N U_{p,n}^*(\bvec k) \psi_{p, \bvec k}^* (\bvec r) \psi_{q, \bvec k'} (\bvec r) U_{q,n'}(\bvec k') e^{i\bvec k \cdot \bvec R -i\bvec k' \cdot \bvec R'}\ud \bvec k \ud \bvec k'\ud \bvec r.\\
\end{eqnarray}

From the orthogonality condition of the Bloch wavefunctions
$$\frac{1}{\vert \Omega^*\vert} \int _{\RR^3} \psi^*_{n', \bvec k'}(\bvec r) \psi_{n, \bvec k}(\bvec r) \ud \bvec r  = \delta_{n', n}\delta (\bvec k'-\bvec k)$$

We have
\begin{eqnarray}
\int_{\RR^3}w_{n, \bvec R}^*(\bvec r)w_{n', \bvec R'}(\bvec r) \ud \bvec r &=& \frac{1}{\vert \Omega^* \vert}\int _{\Omega^*} \ud \bvec k  \sum_{p=1}^N  U_{p,n}^*(\bvec k)  U_{p,n'}(\bvec k) e^{i\bvec k \cdot \bvec R -i\bvec k \cdot \bvec R'}.\\
&=& \frac{1}{\vert \Omega^* \vert}\int _{\Omega^*} \ud \bvec k  \delta_{n, n'} e^{i\bvec k \cdot \bvec R -i\bvec k \cdot \bvec R'}.\\
&=&  \delta_{n, n'} \delta_{\bvec R - \bvec R'}.
\end{eqnarray}

**Solution:**

Derivation for  $S=\langle\psi\vert (\bvec{S}^{\text{tot}})^2\vert\psi\rangle$:

1. Exact

The many body state is spin-pure. It is either a spin-singlet ($S=0$) state or a spin-triplet state ($S=1$).  With proper setup of parameters, the spin-singlet state is the ground state.

2. RHF

The RHF state is given by the Slater determinant of the state $\varphi(\bvec r)\vert{\uparrow}\rangle$ and $\varphi(\bvec r)\vert{\downarrow}\rangle$. This is a spin-singlet state and always has $S=0$.

3. UHF

The UHF state can a Slater determinant generated by $\{\varphi_1(\bvec r)\vert{\uparrow}\rangle, \varphi_2(\bvec r)\vert{\uparrow}\rangle\}$, $\{\varphi_1(\bvec r)\vert{\downarrow}\rangle, \varphi_2(\bvec r)\vert{\downarrow}\rangle\}$, or $\{\varphi_1(\bvec r)\vert{\uparrow}\rangle, \varphi_2(\bvec r)\vert{\downarrow}\rangle\}$.

The first two states require $\varphi_1,\varphi_2$ to be orthogonal and are usually excited states. The ground state is given by
the third one. Compared to RHF, the main difference is that $\varphi_1,\varphi_2$ do not need to be orthogonal, and therefore possibly gives spin-contaminated. Direct computation shows that

$$\langle\psi\vert (\bvec{S}^{\text{tot}})^2\vert\psi\rangle = \frac12 \left(\langle{\uparrow\downarrow}\vert (\bvec{S}^{\text{tot}})^2 \vert {\uparrow\downarrow}\rangle + \langle{\downarrow\uparrow}\vert (\bvec{S}^{\text{tot}})^2 \vert {\downarrow\uparrow}\rangle - 2 \langle{\uparrow\downarrow}\vert (\bvec{S}^{\text{tot}})^2 \vert {\downarrow\uparrow}\rangle \vert\langle \varphi_1\vert\varphi_2\rangle\vert^2 \right).$$

Note that

$$\vert {\uparrow\downarrow}\rangle = \frac{1}{\sqrt{2}}\left(\frac{1}{\sqrt{2}} \vert {\uparrow\downarrow}\rangle + \frac{1}{\sqrt{2}} \vert {\downarrow\uparrow}\rangle  \right) + \frac{1}{\sqrt{2}}\left(\frac{1}{\sqrt{2}} \vert {\uparrow\downarrow}\rangle - \frac{1}{\sqrt{2}} \vert {\downarrow\uparrow}\rangle  \right).$$
The first and second parentheses on the right hand side of the equation are eigenvectors of $\bvec{S}^{\text{tot}})^2$ with eigenvalues $2$ and $0$, respectively.  So

$$\langle\psi\vert (\bvec{S}^{\text{tot}})^2\vert\psi\rangle = 1 - \vert\langle \varphi_1\vert\varphi_2\rangle\vert^2.$$

So the spin of UHF ranges in $[0,1]$, and is spin-contaminated if $\vert\langle \varphi_1\vert\varphi_2\rangle\vert \ne 1$.


```julia
using LinearAlgebra

struct Ham
    R
    rhoa_atom
    rhoa
    Vext
    V_II
    Vext_total
    H_total
end

function init(IntDist,verbose)
    # Nuclei
    println("Distance: ", IntDist)
    R = [-IntDist/2.00,IntDist/2.00]
    rhoa_atom = zeros(Ns,Natoms)
    rhoa = zeros(Ns)
    for j = 1 : Natoms
        dist = gridpos .- R[j]
        @. rhoa_atom[:,j] = - Z[j] / sqrt(2*pi*sigma[j]^2) * (exp(-0.5*(dist/sigma[j])^2 ))
        rhoa .= rhoa + rhoa_atom[:,j]
    end
    # Potential from Nuclei
    Vext = vC_mat * rhoa * dx
    # Nuclear-Nuclear Repulsion

    V_II = vC(IntDist)
    
    if verbose >= 2
        figure(IntDist/dx)
        plot(gridpos,[rhoa_atom vC_mat*rhoa_atom[:,1] vC_mat*rhoa_atom[:,2]])
    end
    
    # Total Hamiltonian
    Vext_total = vC_mat * rhoa[:] * dx
    H_total = Lap + diagm(0=>Vext_total)
    return Ham(R, rhoa_atom, rhoa, Vext, V_II, Vext_total, H_total)
end

function exact(ham::Ham, verbose)
    # Construct the 2-body Hamiltonian (by brute force)
    Iden = Matrix(1.0I,Ns,Ns)
    H_total = ham.H_total
    V_II = ham.V_II
    
    H = kron(H_total,Iden) + kron(Iden,H_total) + diagm(0=>vC_mat[:])
    
    eigH = eigen(H)
    
    # spin treatment
    Psi_nd = reshape(eigH.vectors[:,1],Ns,Ns)
    spin = norm(Psi_nd-Psi_nd')

    # Ground and first excited states
    energies = eigH.values[1:2] .+ V_II
    if verbose >= 1
        println(" Exact energy:", energies, ", <S^2> = ", spin)
    end
    return energies, spin
end

function h2plus(ham::Ham, Nb,verbose)
    H_total = ham.H_total
    V_II = ham.V_II
    
    # Basis Functions are eigenvectors of the one electron system
    Phi = eigvecs(H_total)[:,1:Nb]
    
    # Construct the 2-body Hamiltonian (by brute force)
    Iden = Matrix(1.0I,Ns,Ns)
    
    H = kron(H_total,Iden) + kron(Iden,H_total) + diagm(0=>vC_mat[:])

    Basis = kron(Phi,Phi)
    
    # Project 2-body hamiltonian into basis of one electron eigenvectors
    H_proj = Basis'*(H*Basis)
    
    # Enforce symmetry to make sure ordering of the eigenvalues
    H_proj = 0.5*(H_proj+H_proj')
    
    # Solve eigenvalue problem
    eigH_proj = eigen(H_proj)
    
    # spin treatment
    Psi = Basis * eigH_proj.vectors
    Psi_nd = reshape(Psi[:,1],Ns,Ns)
    spin = norm(Psi_nd-Psi_nd')
    
    # alt method
#     Parity = kron(Iden,Iden)[:,end:-1:1]
#     spin = Psi[:,1]'*Parity*Psi[:,1]
    
    # Ground and first excited states
    energies = eigH_proj.values[1:2] .+ V_II
    if verbose >= 1
        println(" H2+ energy:", energies, ", <S^2> = ", spin)
    end
    return energies, spin
end

function aos(ham::Ham, Nb,verbose)
    rhoa_atom = ham.rhoa_atom
    H_total = ham.H_total
    V_II = ham.V_II
    
    # Make AO Basis
    Phi = zeros(Ns,2*Nb)
    
    Vext_atom = vC_mat * rhoa_atom[:,1] * dx
    H_atom = Lap + diagm(0=>Vext_atom)
    Phi[:,1:Nb] = eigvecs(H_atom)[:,1:Nb]

    Vext_atom = vC_mat * rhoa_atom[:,2] * dx
    H_atom = Lap + diagm(0=>Vext_atom)
    Phi[:,Nb+1:Nb*2] = eigvecs(H_atom)[:,1:Nb]
    
    S_proj = Phi'*Phi
    
    # Construct the 2-body Hamiltonian (by brute force)
    Iden = Matrix(1.0I,Ns,Ns)
    
    H = kron(H_total,Iden) + kron(Iden,H_total) + diagm(0=>vC_mat[:])
    
    Basis_AO = kron(Phi[:,1:Nb],Phi[:,Nb+1:Nb*2])
    
    # Project 2-body hamiltonian into basis of one electron eigenvectors
    H_proj_AO = Basis_AO'*(H*Basis_AO)
    S_proj_AO = Basis_AO'*Basis_AO # orthogonalize since AO basis are not orthogonal
    
    # Enforce symmetry to make sure ordering of the eigenvalues
    H_proj_AO = 0.5*(H_proj_AO+H_proj_AO')    
    
    # Solve eigenvalue problem
    eigH_proj_AO = eigen(H_proj_AO,S_proj_AO)
#     # Total wavefunction eigenvectors
#     Psi_AO = Basis_AO * eigH_proj_AO.vectors
    
    # spin treatment
    Psi = Basis_AO * eigH_proj_AO.vectors
    Psi_nd = reshape(Psi[:,1],Ns,Ns)
    spin = norm(Psi_nd-Psi_nd')
    
    energies = eigH_proj_AO.values[1:2] .+ V_II
    if verbose >= 1
        println(" AO Basis energy:", energies, ", <S^2> = ", spin)
    end
    # Ground and first excited states
    return energies, spin
end

function rhf(ham::Ham, Nb,tol,maxiter,alpha,BasisType,verbose)
    rhoa_atom = ham.rhoa_atom
    Vext_total = ham.Vext_total
    H_total = ham.H_total
    V_II = ham.V_II
    spin = 0 # for RHF


    H_hf = Lap + diagm(0=>(Vext_total))
    # Prepare for SCF
    err = 1.0 ; iter = 0 ; E_hf = 0.0
    while( err > tol && iter < maxiter)
        H_hf = (H_hf+H_hf')/2.0
        eigH_hf = eigen(H_hf)
        
        psi_hf = eigH_hf.vectors[:,1]

        P = 2.0* psi_hf * psi_hf' / dx
        rho = diag(P)
        Vhar = vC_mat * rho * dx
        Hfock = - vC_mat .* P * dx
        
        H_hf_new = Lap + diagm(0=>(Vext_total+0.5*Vhar))
        
        E_hf = 2*(eigH_hf.values[1]) - 0.25 * sum(Vhar .* rho * dx) + V_II
        
        err = norm(H_hf - H_hf_new) / norm(H_hf)
        iter = iter + 1
        if verbose >= 3
            println("   Iter = ", iter, ", err = ", err)
        end
        
        H_hf = (1-alpha) * H_hf + alpha * H_hf_new
    end

    if verbose >= 1
        println(" RHF energy:", E_hf, ", <S^2> = ", spin)
    end
    if verbose >= 2
        println("  Iter = ", iter, ", err = ", err)
    end


    return E_hf, spin
end


function uhf(ham::Ham, Nb,tol,maxiter,alpha,BasisType,verbose)
    rhoa_atom = ham.rhoa_atom
    Vext_total = ham.Vext_total
    H_total = ham.H_total
    V_II = ham.V_II
    
    # Make AO Basis
    Phi = zeros(Ns,2*Nb)
    
    Vext_atom = vC_mat * rhoa_atom[:,1] * dx
    H_atom = Lap + diagm(0=>Vext_atom)
    Phi[:,1:Nb] = eigvecs(H_atom)[:,1:Nb]

    Vext_atom = vC_mat * rhoa_atom[:,2] * dx
    H_atom = Lap + diagm(0=>Vext_atom)
    Phi[:,Nb+1:Nb*2] = eigvecs(H_atom)[:,1:Nb]
    

    ### UHF (Use localized orbitals as initial guess)
    # alpha
    P_a = (Phi[:,1] * Phi[:,1]')/ dx
    rho_a = diag(P_a)
    Hfock_a = - vC_mat .* P_a * dx
    # beta
    P_b = (Phi[:,Nb+1] * Phi[:,Nb+1]')/ dx
    rho_b = diag(P_b)
    Hfock_b = - vC_mat .* P_b * dx
    # total
    rho_total = rho_a + rho_b
    Vhar = vC_mat * rho_total * dx
    
    # hamiltonians
    H_hf_a = Lap + diagm(0=>(Vext_total+Vhar)) + Hfock_a
    H_hf_b = Lap + diagm(0=>(Vext_total+Vhar)) + Hfock_b
    
    spin = 0.0
    # Prepare for SCF
    err = 1.0 ; iter = 0 ; E_hf = 0.0
    while( err > tol && iter < maxiter)
        # alpha
        H_hf_a = (H_hf_a + H_hf_a') * 0.5
        eigH_hf_a = eigen(H_hf_a) 
        psi_hf_a = eigH_hf_a.vectors[:,1]
        P_a = psi_hf_a * psi_hf_a' / dx
        rho_a = diag(P_a)
        Hfock_a = - vC_mat .* P_a * dx
        
        # beta
        H_hf_b = (H_hf_b + H_hf_b') * 0.5
        
        eigH_hf_b = eigen(H_hf_b) 
        psi_hf_b = eigH_hf_b.vectors[:,1]
        P_b = psi_hf_b * psi_hf_b' / dx
        rho_b = diag(P_b)
        Hfock_b = - vC_mat .* P_b * dx
        
        #total
        rho_total = rho_a + rho_b
        Vhar = vC_mat * rho_total * dx

        # hamiltonians
        H_hf_new_a = Lap + diagm(0=>(Vext_total+Vhar)) + Hfock_a
        H_hf_new_b = Lap + diagm(0=>(Vext_total+Vhar)) + Hfock_b
        
        # spin 
        rho_spin = rho_a - rho_b
        spin = 1.0 - abs(psi_hf_a'*psi_hf_b)^2 # note that dx^2 is absored in the normalization condition of psi_hf
        
        E_hf = real(eigH_hf_a.values[1]+eigH_hf_b.values[1] - 0.50 * sum(Vhar .* rho_total * dx) -
            0.5 * tr(psi_hf_a'*(Hfock_a*psi_hf_a)) - 0.5 * tr(psi_hf_b'*(Hfock_b*psi_hf_b)) + V_II)
        
        err = norm(H_hf_a - H_hf_new_a) / norm(H_hf_a) + norm(H_hf_b - H_hf_new_b) / norm(H_hf_b)
        iter = iter + 1
        if verbose >= 3
            println("   Iter = ", iter, ", err = ", err)
        end
        #update
        H_hf_a = (1-alpha) * H_hf_a + alpha * H_hf_new_a
        H_hf_b = (1-alpha) * H_hf_b + alpha * H_hf_new_b
    end

    
    if verbose >= 1
        println(" UHF energy:", E_hf, ", <S^2> = ", spin)
    end
    if verbose >= 2
        println("  Iter = ", iter, ", err = ", err)
    end
    return E_hf, spin
end
```

```julia
using LinearAlgebra
using PyPlot

# quick and dirty implementation
function everything(a,dx)
    #discretization of position space
    κ = 0.100
    ϵ_0 = 1.0

    # number of atoms in super cell
    Natoms = 10
    @assert(mod(Natoms,2) == 0)

    #length of supercell
    Ls = Natoms*a
    #list of positions of atoms in supercell
    R = zeros(1, Natoms)

    for atom in 0:1:Natoms-1
        R[atom+1]=-Ls/2 + (a/2) + a*(atom)
    end    

    #charge of nucleus for each atom in list of atoms
    Z = 1.0 * ones(Natoms)
    sigma = 0.5 * ones(Natoms)

    # 1/r
    vC(x) = 1.0 / (ϵ_0 * sqrt(x^2 + κ^2))
    # distance to point with periodic boundary condition
    minx(x,y) = abs(y-x) - round(abs(y-x)/Ls)*Ls

    # real space grid
    gridpos = collect(range(-Ls/2,stop=Ls/2,step=dx))
    Ns = length(gridpos)

    vC_mat = [vC( minx(gridpos[i],gridpos[j]) ) for i in 1:Ns, j in 1:Ns]

    rhoa_atom = zeros(Ns,Natoms)
    rhoa = zeros(Ns)

    for j = 1 : Natoms
        dist = gridpos .- R[j]
        @. rhoa_atom[:,j] = - Z[j] / sqrt(2*pi*sigma[j]^2) * (exp(-0.5*(dist/sigma[j])^2 ))
        rhoa .= rhoa + rhoa_atom[:,j]
    end

    Vext = vC_mat * rhoa * dx 

    Lap = 1.0 / (2.0*dx^2) * diagm(-1=>-ones(Ns-1),0=>2.0*ones(Ns),1=>-ones(Ns-1))

    Lap[1,Ns]=Lap[1,2]
    Lap[Ns,1]=Lap[2,1]

    #ion ion potential energy
    V_II=0

    for atom_1_position in R
        for atom_2_position in R
            if atom_1_position < atom_2_position
                V_II += vC(minx(atom_2_position, atom_1_position))
            end
        end
    end

    # HF ground state
    Nocc = div(Natoms, 2)
    tol = 1e-5
    
    rho = zeros(Ns)
    err = 1.0
    iter = 0
    alpha = 0.1
    E_hf = 0.0
    maxiter = 1000
    while( err > tol && iter < maxiter )
        #Vnuclei= vC_mat * rhoa * dx
        Vhar =  0.5*vC_mat * rho * dx 
        H_hf = Lap + diagm(0=>(Vext+Vhar))
        H_hf = (H_hf+H_hf')*0.5

        eigH_hf = eigen(H_hf)
        psi_hf = eigH_hf.vectors[:,1:Nocc]
        #display(psi_hf )
        rhonew = vec(sum(2*abs.(psi_hf).^2,dims=2)/ dx)
        E_hf = 2 * sum(eigH_hf.values[1:Nocc]) -  0.5*sum(Vhar .* rho * dx)  + V_II 

        err = norm(rho - rhonew) / norm(rho)
        iter = iter + 1
        #println("Iter = ", iter,", E_hf = ", E_hf , ", err = ", err)
        rho = (1-alpha) * rho + alpha * rhonew
    end
    #println("Hartree Fock Calculation Converged for length ", string(a))
    #println("Iter = ", iter, ", err = ", err)

    return E_hf
end

Npoint=20
dx_vec = LinRange(0.05,0.25,Npoint)
energy_vec = zeros(Npoint)

for l = 1 : Npoint
    println("dx=", dx_vec[l])
    energy_vec[l] = everything(1.3,dx_vec[l])
end

using PyPlot

figure(1)
plot(dx_vec,energy_vec)
xlabel("value of dx")
ylabel("E(dx)")
```

```julia
# Density matrix method, use McWeeny's purification

function make_density_matrix!(H,mu,M)
    a = maximum(sum(abs.(H),dims=2))
    # println(a)
    if mu < 0.0
        k = 0.5/(a-mu)
        r = 1 - k*a
    else
        k = 0.5/(mu+a)
        r = k*a
    end
    D = k*H + r*I
    # results=eigen(D)
    # println(results.values)
    for iter = 1:M
        D = 3*D^2 - 2*D^3
        D = 0.5*(D+D')
    end
    return I-D
end


function make_density_matrix(H,nev,M)
    l1_norm = maximum(sum(abs.(H),dims=2))
    a = -l1_norm
    b = l1_norm
    mid = (a+b)/2
    D = make_density_matrix!(H,mid,M)
    nev_temp = tr(D)
    # println(nev_temp)
    while abs(nev_temp-nev) > 0.1
        # println(nev_temp,"!")
        # println(a," ",mid," ",b)
        if nev_temp > nev + 0.1
            b = mid
            mid = (a+b)/2
            D = make_density_matrix!(H,mid,M)
            nev_temp = tr(D)
        elseif nev_temp < nev - 0.1
            a = mid
            mid = (a+b)/2
            D = make_density_matrix!(H,mid,M)
            nev_temp = tr(D)
        else
            return D
        end
    end
    return D
end

# quick and dirty implementation
function everything_dm(a,dx)
    #discretization of position space
    κ = 0.100
    ϵ_0 = 1.0

    # number of atoms in super cell
    Natoms = 10
    @assert(mod(Natoms,2) == 0)

    #length of supercell
    Ls = Natoms*a
    #list of positions of atoms in supercell
    R = zeros(1, Natoms)

    for atom in 0:1:Natoms-1
        R[atom+1]=-Ls/2 + (a/2) + a*(atom)
    end    

    #charge of nucleus for each atom in list of atoms
    Z = 1.0 * ones(Natoms)
    sigma = 0.5 * ones(Natoms)

    # 1/r
    vC(x) = 1.0 / (ϵ_0 * sqrt(x^2 + κ^2))
    # distance to point with periodic boundary condition
    minx(x,y) = abs(y-x) - round(abs(y-x)/Ls)*Ls

    # real space grid
    gridpos = collect(range(-Ls/2,stop=Ls/2,step=dx))
    Ns = length(gridpos)

    vC_mat = [vC( minx(gridpos[i],gridpos[j]) ) for i in 1:Ns, j in 1:Ns]

    rhoa_atom = zeros(Ns,Natoms)
    rhoa = zeros(Ns)

    for j = 1 : Natoms
        dist = gridpos .- R[j]
        @. rhoa_atom[:,j] = - Z[j] / sqrt(2*pi*sigma[j]^2) * (exp(-0.5*(dist/sigma[j])^2 ))
        rhoa .= rhoa + rhoa_atom[:,j]
    end

    Vext = vC_mat * rhoa * dx 

    Lap = 1.0 / (2.0*dx^2) * diagm(-1=>-ones(Ns-1),0=>2.0*ones(Ns),1=>-ones(Ns-1))

    Lap[1,Ns]=Lap[1,2]
    Lap[Ns,1]=Lap[2,1]

    #ion ion potential energy
    V_II=0

    for atom_1_position in R
        for atom_2_position in R
            if atom_1_position < atom_2_position
                V_II += vC(minx(atom_2_position, atom_1_position))
            end
        end
    end

    # HF ground state
    Nocc = div(Natoms, 2)
    tol = 1e-5
    
    rho = zeros(Ns)
    err = 1.0
    iter = 0
    alpha = 0.1
    E_hf = 0.0
    maxiter = 1000
    while( err > tol && iter < maxiter )
        #Vnuclei= vC_mat * rhoa * dx
        Vhar =  0.5*vC_mat * rho * dx 
        H_hf = Lap + diagm(0=>(Vext+Vhar))
        H_hf = (H_hf+H_hf')*0.5

        dm = 2*make_density_matrix(H_hf,Nocc,40)
        rhonew = diag(dm)/dx
        ev_sum = sum(dm.*H_hf)
        E_hf = ev_sum -  0.5*sum(Vhar .* rho * dx)  + V_II 

        err = norm(rho - rhonew) / norm(rho)
        iter = iter + 1
        #println("Iter = ", iter,", E_hf = ", E_hf , ", err = ", err)
        rho = (1-alpha) * rho + alpha * rhonew
    end
    #println("Hartree Fock Calculation Converged for length ", string(a))
    #println("Iter = ", iter, ", err = ", err)

    return E_hf
end

energy_diag = everything_dm(0.51,0.05)
energy_dm = everything_dm(0.51,0.05)

println("Energy by diagonalization = ", energy_diag)
println("Energy by purification    = ", energy_dm)
```

**Hint:** The shape of the density should be responsible for the [$\pi$ bonds](https://www.masterorganicchemistry.com/2017/05/05/the-pi-molecular-orbitals-of-benzene/) of the benzene molecule.


```julia
using LinearAlgebra
using SparseArrays
using Arpack # for eigs
using IterativeSolvers # for cg
using SpecialFunctions # for erf
using PyPlot


function DFT_hydrogen_mol(g,L,D,tol,maxit,verbose)
    g3 = g^3
    p = collect(range(-L,stop=L,length=g))
    Pts = [[p[i],p[j],p[k]] for i=1:g,j=1:g,k=1:g]
    Pts = reshape(Pts,(g3))
    X = [Pts[i][1] for i=1:g3]
    Y = [Pts[i][2] for i=1:g3]
    Z = [Pts[i][3] for i=1:g3]
    h = p[2]-p[1]
    R = sqrt.(X.^2+Y.^2+Z.^2)
    R1 = sqrt.((X.-D/2).^2+Y.^2+Z.^2)
    R2 = sqrt.((X.+D/2).^2+Y.^2+Z.^2)
    Vext = -1.0./R1 - 1.0./R2
    L = spdiagm(0=>-2*ones(g),1=>ones(g-1),-1=>ones(g-1))/(h^2)
    Id = SparseMatrixCSC{Float64}(I,g,g)
    L3 = kron(kron(L,Id),Id) + kron(kron(Id,L),Id) + kron(kron(Id,Id),L)

    Vtot = Vext
    # Use compensation charge to handle long range Coulomb potential
    ncomp = exp.(-R1.^2/2) + exp.(-R2.^2/2)
    ncomp = -2.0*ncomp/sum(ncomp)/h^3
    ncomppot = -1.0./R1.*erf.(R1/sqrt(2)) - 1.0./R1.*erf.(R2/sqrt(2))
    iter = 1
    Etot = 0
    dE = 1.0+tol
    E=0;T=0;Eext=0;Eh=0;Ex=0;Enuc=0
    while iter<maxit && dE > tol
        # global Vtot, iter, dE, Etot
        # global E,T,Eext,Eh,Ex,Enuc
        (ev,psi,nconv,niter,nmult,resid) = eigs(-0.5*L3+spdiagm(0=>Vtot),nev=1,which=:SR,ritzvec=true)
        E = ev[1]
        psi /= h^1.5
        n = dropdims( 2*psi.^2, dims=2 )
        Vx = -(3/pi)^(1/3)*n.^(1/3)  # exchange correlation
        # solve L3 Vh = -4*pi*n to get Hartree potential
        Vh = cg(L3,-4*pi*(n+1.0*ncomp);tol=1e-7,maxiter=400)-1.0*ncomppot  # solve a Poisson's eq using CG
        Vtot = 1.0*Vx + Vh + Vext
        T = 2*sum(dot(psi,(-0.5*L3)*psi))*h^3
        Eext = sum(n.*Vext)*h^3
        Eh = 0.5*sum(n.*Vh)*h^3
        Ex = sum(-(3/4)*(3/pi)^(1/3)*n.^(4/3))*h^3
        Enuc = 1.0/D
        dE = T + Eext + Eh + 1.0*Ex + Enuc - Etot
        Etot += dE
        dE = abs(dE)
        if verbose >= 2
            println("------------- Iteration ",iter," ----------------")
            println("Eigenvalue ",E)
            println("Kinetic energy ",T)
            println("External energy ",Eext)
            println("Hartree energy ", Eh)
            println("Exchange correlation energy ", Ex)
            println("Nuclei repulsion energy ", Enuc)
            println("Total energy ", Etot)
        end
        iter += 1
    end
    if verbose >= 1
        println("------------- Converged result ----------------")
        println("Eigenvalue ",E)
        println("Kinetic energy ",T)
        println("External energy ",Eext)
        println("Hartree energy ", Eh)
        println("Exchange correlation energy ", Ex)
        println("Nuclei repulsion energy ", Enuc)
        println("Total energy ", Etot)
    end
#     println("Number of iterations: ",iter)
    return Etot
end
```

```julia
let
    L = 3
    g = 30  # takes too long to run, can be changed to g=30
    verbose = 0
    tol = 1e-3
    maxit = 100
    D_list = [0.5,0.7,1.0,1.4,1.5,2.0,2.5,3.0,3.5,4.0]  # list of bond lengths
    Etot_list = fill(0.0,length(D_list))
    for ix = 1:length(D_list)
        println("D = ", D_list[ix])
        Etot_list[ix] = DFT_hydrogen_mol(g,L,D_list[ix],tol,maxit,verbose)
    end
    plot(D_list,Etot_list)
    xlabel(L"Bond length $R$")
    ylabel(L"$E_{tot}$");
end
```
